# Train/Fit & Eval Data

```python
from pathlib import Path
from methods import (
    MKQAExportSpec,
    export_mkqa_to_jsonl,
    project_data_paths,
    write_dataset_registry,
)

ROOT = Path("/media/am/AM/mi-lens")
PATHS = project_data_paths(ROOT)

mkqa_da_trainfit = export_mkqa_to_jsonl(
    MKQAExportSpec(
        dataset_name="mkqa",
        split="train",
        question_language="da",
        answer_language="da",
        start_idx=0,
        max_records=5000,
        output_path=PATHS.train_fit_dir / "mkqa_da_trainfit_5000.jsonl",
    )
)

mkqa_en_trainfit = export_mkqa_to_jsonl(
    MKQAExportSpec(
        dataset_name="mkqa",
        split="train",
        question_language="en",
        answer_language="en",
        start_idx=0,
        max_records=5000,
        output_path=PATHS.train_fit_dir / "mkqa_en_trainfit_5000.jsonl",
    )
)

mkqa_da_eval = export_mkqa_to_jsonl(
    MKQAExportSpec(
        dataset_name="mkqa",
        split="train",
        question_language="da",
        answer_language="da",
        start_idx=5000,
        max_records=5000,
        output_path=PATHS.eval_dir / "mkqa_da_eval_5000.jsonl",
    )
)

mkqa_en_eval = export_mkqa_to_jsonl(
    MKQAExportSpec(
        dataset_name="mkqa",
        split="train",
        question_language="en",
        answer_language="en",
        start_idx=5000,
        max_records=5000,
        output_path=PATHS.eval_dir / "mkqa_en_eval_5000.jsonl",
    )
)

write_dataset_registry(
    PATHS.registry_path,
    [
        mkqa_da_trainfit,
        mkqa_en_trainfit,
        mkqa_da_eval,
        mkqa_en_eval,
    ],
    note="Fixed unshuffled MKQA train_fit/eval slices with matched DA/EN ranges.",
)
```
